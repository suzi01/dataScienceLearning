```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
destination_df = pd.read_csv('destinations.csv')
destination_df
```

1.How many rows and columns are there in your file?

```python
destination_df.shape
```

2. Print row 3-8 ( using iloc/loc).

```python
rows_df = destination_df.loc[3:8]
rows_df
```

3.Find the mean number of all-inclusive hotels across all destinations.

```python
average_of_allInclusive_hotels = destination_df['No_of _inclusive_hotels'].mean()
average_of_allInclusive_hotels
```

4.Find the lowest scoring destination.

```python
lowest_scoring_destination = destination_df.sort_values('Feedback_score', ascending = True)
lowest_scoring_destination
```

```python
lowest_scoring_destination.head(1)
```

5. Find the highest scoring destination.

```python
highest_scoring_destination = lowest_scoring_destination.tail(1)
highest_scoring_destination
```

6.Find all the destinations where there are more than 9 all-inclusive hotels.

```python
newDes_df = destination_df[destination_df['No_of _inclusive_hotels'] > 9][['Destination ']]
newDes_df
```

7.Filter the data by destination and score above 8.  

```python
score_df = destination_df[destination_df['Feedback_score'] > 9][['Destination ']]
score_df
```

8. Filter the data by destination and score below 2 ( I need to know if these destinations should be removed or there is a problem)

9. Is there a correlation between number of all-inclusive hotels and score?

```python
correlation = destination_df['No_of _inclusive_hotels'].corr(destination_df['Feedback_score'])

correlation
```

The correlation measured is postive but very small and close to zero. Therfore it can be assumed that there is a very weak relationship between number of all-cinlsive hotels and feedback score.

```python

fig, axes = plt.subplots(3, 1, figsize=(16, 19))
plt.tight_layout(pad=13)

# Use the axes for plotting
axes[0].set_title("Feedback score for each destination using simple Seaborn bar plot.")
axes[0].set_xticklabels(destination_df['Destination '],rotation=90)
sns.barplot('Destination ', 'Feedback_score', data = destination_df, ax = axes[0] )

axes[1].set_title("Feedback score for each destination using a line and bar plot with Matplotlib")
axes[1].bar(destination_df['Destination '], destination_df['Feedback_score'])
axes[1].set_xlabel('Destinations')
axes[1].set_ylabel('Feedback Score provided')
axes[1].set_xticklabels(destination_df['Destination '],rotation=90)
axes[1].plot(destination_df['Destination '], destination_df['Feedback_score'], 'o--r')


axes[2].set_title('Feedback score for each destination using a scatterplot')
axes[2].set_xticklabels(destination_df['Destination '],rotation=90)
sns.scatterplot(destination_df['Destination '], 
                destination_df['Feedback_score'],
                s=100);

```
